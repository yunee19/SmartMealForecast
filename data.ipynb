```python
# Import thư viện / Import pandas library to handle data
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv('food_choices.csv')

# Hiển thị 5 dòng đầu tiên của dữ liệu để kiểm tra/ Display the first 5 rows to inspect the data
df.head()
```

```python
# Kiểm tra thông tin của DataFrame
df.info()

# Kiểm tra các giá trị bị thiếu (null values)
df.isnull().sum()
```

```python
# Thay thế giá trị NaN trong cột '본사휴가자수' bằng giá trị trung bình của cột này
df['본사휴가자수'].fillna(df['본사휴가자수'].mean(), inplace=True)

# Hoặc xóa các dòng có giá trị thiếu
df.dropna(inplace=True)
```

```python
# Chuyển đổi cột '일자' thành kiểu datetime
df['일자'] = pd.to_datetime(df['일자'])
```

```python
df.describe()
```

```python
df_filtered = df[df['본사휴가자수'] > 100]
df_filtered.head()
```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ phân phối của cột '본사정원수'
plt.hist(df['본사정원수'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of 본사정원수')
plt.xlabel('본사정원수')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.plot(df['일자'], df['본사휴가자수'])
plt.title('휴가자수 (People on Leave) Over Time')
plt.xlabel('Date')
plt.ylabel('People on Leave')
plt.xticks(rotation=45)
plt.show()
```

```python
df.to_csv('processed_food_choices.csv', index=False)
```
